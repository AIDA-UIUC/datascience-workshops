```python
# Using sample data
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
df = pd.DataFrame()
df['y'] = y
df['x'] = x
fig = px.scatter(df, x = "x", y = "y")
fig.show()
```
