```python
gpa = pd.read_csv("https://github.com/wadefagen/datasets/raw/master/gpa/uiuc-gpa-dataset.csv")
print(gpa)
```

```python
# There's probably a much better way to do this! See if you can figure it out...
gpa = pd.read_csv("https://github.com/wadefagen/datasets/raw/master/gpa/uiuc-gpa-dataset.csv")

gpa['total_students'] = gpa['A+'] + gpa['A'] + gpa['A-'] + gpa['B'] + gpa['B+'] + gpa['B-'] + gpa['C+'] + gpa['C'] + gpa['C-'] + gpa['D+'] + gpa['D'] + gpa['D-'] + gpa['F']

gpa['GPA'] = (gpa['A+'] * 4 + gpa['A'] * 4 + gpa['A-'] * 3.67 + gpa['B'] * 3 + gpa['B+'] * 3.33 + gpa['B-'] * 2.67 + gpa['C+'] * 2.33 + gpa['C'] * 2 + gpa['C-'] * 1.67 + gpa['D+'] * 1.33 + gpa['D'] + gpa['D-'] * 0.67) / gpa['total_students']
gpa["4s given"] = (gpa['A'] + gpa['A+']) / gpa['total_students']

print(gpa)
```

```python
gpa_cs = gpa[gpa['Subject'] == 'CS']
gpa_cs
```

```python
fig = px.scatter(x = gpa_cs['4s given'], y = gpa_cs['GPA'], size = gpa_cs['total_students'], hover_name=gpa_cs["Course Title"])
fig.show()
```

```python
fig.write_html("gpa_cs.html")
```

```python
gpa_ece = gpa[gpa['Subject'] == 'ECE']
print(gpa_ece)
```

```python
gpa_abe = gpa[gpa['Subject'] == "ABE"]
gpa_ae = gpa[gpa['Subject'] == "AE"]
gpa_me = gpa[gpa['Subject'] == "ME"]
gpa_bioe = gpa[gpa['Subject'] == "BIOE"]
gpa_chbe = gpa[gpa['Subject'] == "CHBE"]
gpa_npre = gpa[gpa['Subject'] == "NPRE"]
gpa_mse = gpa[gpa['Subject'] == "MSE"]
gpa_cee = gpa[gpa['Subject'] == "CEE"]
gpa_ise = gpa[gpa['Subject'] == "IE"]

gpa_engr = pd.concat([gpa_cs, gpa_ece, gpa_abe, gpa_ae, gpa_me, gpa_chbe, gpa_bioe, gpa_npre, gpa_mse, gpa_cee, gpa_ise])
print(gpa_engr)
```

```python
fig = px.scatter(gpa_engr, x = '4s given', y = 'GPA', size = gpa_engr['total_students'], color = gpa_engr['Subject'])
fig.show()
```

```python
gpa_engr_recent = gpa_engr[gpa_engr['Year'] == 2019]

print(gpa_engr_recent)
```

```python
fig_recent = px.scatter(gpa_engr_recent, x = '4s given', y = 'GPA', size = gpa_engr_recent['total_students'], color = gpa_engr_recent['Subject'], hover_name= "Course Title")
fig_recent.show()
```

## Problem 2: wadefagan's Useful Datasets: [Perception-of-Probability-Words](https://github.com/wadefagen/datasets/tree/master/Perception-of-Probability-Words)
Some helpful links we used while doing the same:
- https://plot.ly/python/
- https://plot.ly/python/box-plots/
- https://plot.ly/python/v3/axes/
- https://plot.ly/python/click-events/

```python
perception = pd.read_csv("https://raw.githubusercontent.com/wadefagen/datasets/master/Perception-of-Probability-Words/survey-results.csv")
perception
```

```python
perception_dropped = perception.drop(["Your Highest Level of Education", "Your Gender Identity", "Age Range"], axis = 1)
perception_dropped
```

```python
perception_dropped.mean().sort_values()
sorted_perceptions =  perception_dropped.reindex(perception_dropped.mean().sort_values().index, axis=1)
sorted_perceptions
```

```python
fig = go.Figure()
for col_name in sorted_perceptions.columns:
    fig.add_trace(go.Box(
        x = sorted_perceptions[col_name],
        y = [str(col_name)] * sorted_perceptions.shape[0],
        name = str(col_name),
        orientation = "h"
    ))

fig.update_layout(
  hovermode = "closest",
  showlegend = True,
  title = "Perception of Probability Words",
  xaxis = {
    'title': "Probability"
  },
  yaxis = {
    'showline': True,
    'title': "Perception"
  }
)

fig.show()
```

## Problem 3: Visualizing the Stock Market
If you are looking to grab some stock market data, websites like [Yahoo Finance](https://finance.yahoo.com/quote/MSFT/history?p=MSFT) keep clean records of most stocks. Developed libraries like plot.ly also do have public datasets of their own that you can download and use anytime: [here!](https://github.com/plotly/datasets)

Use this dataset to recreate a Time Series plot with Microsoft's stock prices since January of 2010!

This is a sneak peek at some cooler type of graphs we'll be doing next week. The plotly python examples website did something _very_ similar if you need some [help](https://plotly.com/python/time-series/)!

```python
msft = pd.read_csv("https://raw.githubusercontent.com/AIDA-UIUC/datascience-workshops/master/datasets/MSFT.csv")
msft
```

```python
# first row will be Nan, but that's okay
# calculate the difference between today's High and yesterday's High
# we didn't talk about .shift() during last week's pandas workshop but it allows rows in dataframes to access their previous or next neighbors!

msft['change_high'] = msft['High'] - msft['High'].shift()
msft
```

```python
fig = px.line(msft, x='Date', y='change_high', title='Microsfot Time Series with Range Slider and Selectors!', hover_name = "Date")

fig.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=1, label="YTD", step="year", stepmode="todate"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(count=5, label="5y", step="year", stepmode="backward"),
            dict(count=2, label="2 years", step="year", stepmode='backward'),
            dict(step="all")
        ])
    )
)

fig.show()
```
