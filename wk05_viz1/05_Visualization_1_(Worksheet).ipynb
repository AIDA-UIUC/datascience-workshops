```python
# Using sample data
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
fig = go.Figure()
fig.add_trace(go.Scatter(x = x, y = y, mode = 'markers'))
fig.add_trace(go.Scatter(x = x, y = np.cos(x), mode = 'markers'))
fig.show()
```
